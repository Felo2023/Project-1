```python
import pandas as pd
import matplotlib as plt
from pathlib import Path
```

```python
#dataframes to decode main tables
entidades=pd.read_csv('../Data/ENTIDADES.csv')
nacionalidad=pd.read_csv('../Data/NACIONALIDAD.csv')
paciente=pd.read_csv('../Data/PACIENTE.csv')
result_antig=pd.read_csv('../Data/RESULT_ANTIG.csv')
result_lab=pd.read_csv('../Data/RESULT_LAB.csv')
sector=pd.read_csv('../Data/SECTOR.csv')
sexo=pd.read_csv('../Data/SEXO.csv')
si_no_pd=pd.read_csv('../Data/SI_NO.csv')
```

```python
dataframes=[]
for i in range(1):
    
    datalocation=f'../Data/Covid202{i}-MUERTOS.csv'
    dataframes.append(pd.read_csv(datalocation).drop({'FECHA_ACTUALIZACION',
                                                'ENTIDAD_NAC',
                                                'ENTIDAD_RES',
                                                'MUNICIPIO_RES',
                                                'ORIGEN',
                                                'NEUMONIA',
                                                'HABLA_LENGUA_INDIG',
                                                'OTRA_COM',
                                                'TABAQUISMO',
                                                'OTRO_CASO',
                                                'TOMA_MUESTRA_LAB',
                                                'TOMA_MUESTRA_ANTIGENO',
                                                'MIGRANTE',
                                                'PAIS_NACIONALIDAD'}
                                                ,axis=1))
    print(datalocation)
```

```python
complete_data=pd.DataFrame()
```

```python
for dataframe in dataframes:
    complete_data=pd.concat([complete_data,dataframe])
```

```python
#Replace data on dataframe (EXCEPT SI_NO)
complete_data2=pd.DataFrame.merge(complete_data,entidades,on='ENTIDAD_UM',how='inner')
complete_data2=complete_data2.drop({'ENTIDAD_UM'},axis=1)

complete_data2=pd.DataFrame.merge(complete_data2,paciente,on='TIPO_PACIENTE',how='inner')
complete_data2=complete_data2.drop({'TIPO_PACIENTE'},axis=1)

complete_data2=pd.DataFrame.merge(complete_data2,sector,on='SECTOR',how='inner')
complete_data2=complete_data2.drop({'SECTOR'},axis=1)

complete_data2=pd.DataFrame.merge(complete_data2,sexo,on='SEXO',how='inner')
complete_data2=complete_data2.drop({'SEXO'},axis=1)

complete_data2=complete_data2.loc[complete_data2['NACIONALIDAD']==1]
complete_data2=complete_data2.drop({'NACIONALIDAD'},axis=1)

complete_data2=complete_data2.loc[complete_data2['INDIGENA']==2]
complete_data2=complete_data2.drop({'INDIGENA'},axis=1)

complete_data2=complete_data2.loc[complete_data2['FECHA_DEF']!='9999-99-99']
complete_data2=complete_data2.drop({'FECHA_DEF'},axis=1)
```

```python

si_no_pd=si_no_pd.rename(columns={'CLAVE':'INTUBADO'})
complete_data2=pd.DataFrame.merge(complete_data2,si_no_pd,on='INTUBADO',how='inner')
complete_data2=complete_data2.drop({'INTUBADO'},axis=1)
complete_data2=complete_data2.rename(columns={'DESCRIPCION_y':'INTUBADO'})
si_no_pd=si_no_pd.rename(columns={'INTUBADO':'CLAVE'})
```
